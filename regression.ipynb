```python
test_data = pd.read_csv('test.csv')
train_data = pd.read_csv('train.csv')
train_x = np.array(train_data.x, int)
train_y = np.array(train_data.y, float)
test_x = np.array(test_data.x, int)
test_y = np.array(test_data.y, float)
```

Defining a linear model and shaping the dataset

```python
def linear_model(w, x, b):
    return np.dot(w, x) + b

def append_ones(x_s):
    return np.hstack((np.reshape(x_s, (x_s.shape[0],1)), np.ones((x_s.shape[0],1),)))
```

```python
new_train_x = append_ones(train_x)
new_train_y = append_ones(train_y)
new_test_x = append_ones(test_x)
new_test_y = append_ones(test_y)
```

Computing optimised weight

```python
def compute_weight(Xs, Ys):
    mul1 = np.linalg.inv(np.matmul(np.transpose(Xs), Xs))
    mul2 = np.matmul(np.transpose(Xs), Ys)
    return np.matmul(mul1, mul2)
```

```python
w = compute_weight(new_train_x, new_train_y)
print(f'bias: {w[0][1]:.50f}')
print(f'optmised weight: {w[1][1]:.5f}')
```
